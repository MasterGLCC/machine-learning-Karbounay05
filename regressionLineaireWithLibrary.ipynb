```python
# Import des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  # modèle de régression linéaire
```

```python
# 1. Données du dataset (Heures d'étude et Notes finales)
heures_etude = np.array([2, 4, 3, 8, 10, 9, 3, 7])  # matrice X (variable explicative)
notes_finales = np.array([10, 10, 11, 15, 17, 18.5, 11, 16])  # vecteur Y (cible)
```

```python
# 2. Transformation de Y en vecteur colonne (optionnel pour sklearn mais gardé pour cohérence)
Y = notes_finales.reshape(-1, 1)
```

```python
# 3. Transformation de X au format attendu par sklearn (matrice 2D)
X = heures_etude.reshape(-1, 1)
```

```python
# 4. Création du modèle de régression linéaire
model = LinearRegression()
```

```python
# 5. Entraînement du modèle sur les données (apprentissage)
model.fit(X, Y)

# 6. Extraction des paramètres du modèle
a = model.coef_[0][0]  # coefficient directeur (pente)
b = model.intercept_[0]  # ordonnée à l'origine (biais)

print(f"Résultat du modèle sklearn : f(x) = {a:.2f} * x + {b:.2f}")
```

```python
# 7. Création des points pour tracer la droite de régression
X_new = np.array([[0], [10]])  # valeurs de 0 à 10 heures pour tracer la droite
y_predict = model.predict(X_new)  # prédictions du modèle
```

```python
# 8. Affichage des points réels
plt.scatter(heures_etude, notes_finales, color='red', label='Données réelles')

# 9. Tracé de la droite de régression
plt.plot(X_new, y_predict, "b-", label="Modèle sklearn (Régression)")

# 10. Personnalisation du graphique
plt.xlabel("Heures d'étude")  # label axe X
plt.ylabel("Note finale /20")  # label axe Y
plt.title("Régression linéaire avec sklearn")  # titre du graphique
plt.legend()  # affichage de la légende
plt.grid(True, linestyle='--', alpha=0.6)  # grille pour lisibilité

# 11. Affichage du graphique
plt.show()
```

```python
# 12. Saisie utilisateur pour prédiction
x_manuel = float(input("Entrez le nombre d'heures d'étude : "))

# 13. Prédiction pour une nouvelle valeur
y_manuel = model.predict(np.array([[x_manuel]]))[0][0]

print(f"La note prédite pour {x_manuel}h est : {y_manuel:.2f}/20")

# 14. Ajout du point utilisateur dans le graphique
plt.scatter(heures_etude, notes_finales, color='red', label='Données réelles')
plt.plot(X_new, y_predict, "b-", label="Droite sklearn")
plt.scatter(x_manuel, y_manuel, color='green', s=150, marker='*',
            label=f'Prédiction ({y_manuel:.2f})')

# 15. Annotation du point prédit
plt.annotate(f"{y_manuel:.2f}", (x_manuel, y_manuel),
             color='green', fontweight='bold')

# 16. Configuration finale du graphique
plt.xlabel("Heures d'étude")
plt.ylabel("Note finale /20")
plt.title("Prédiction avec sklearn")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# 17. Affichage final
plt.show()
```
