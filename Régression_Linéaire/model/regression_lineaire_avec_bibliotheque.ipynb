```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Chargement des données
df = pd.read_csv("../dataset/etudiant.csv")

# Variables
X = df[['Heures_etude']]
y = df['Note']

# Création et entraînement du modèle
model = LinearRegression()
model.fit(X, y)

# Paramètres du modèle
a = model.coef_[0]
b = model.intercept_

# Prédictions
y_pred = model.predict(X)

# Évaluation
mse = mean_squared_error(y, y_pred)
print(f"Pente (a): {a:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"MSE: {mse:.2f}")

# ===== VISUALISATION (UN SEUL GRAPHE) =====
pt.figure()

# Données réelles
pt.scatter(X, y, color='blue', label="Données réelles")

# Droite de régression
pt.plot(X, y_pred, color='green', label="Régression linéaire")

# Mise en forme
pt.title("Modèle de Régression Linéaire")
pt.xlabel("Heures d'étude")
pt.ylabel("Note")
pt.legend()

pt.show()
```
