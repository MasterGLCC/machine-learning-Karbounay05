```python
import pandas as pd
import matplotlib.pyplot as pt  # alias personnalisé
import numpy as np

# Chargement des données
df = pd.read_csv("../dataset/etudiant.csv")
X = df['Heures_etude'].values
y = df['Note'].values

# Calcul des moyennes
mean_x = np.mean(X)  # moyenne de X
mean_y = np.mean(y)  # moyenne de y

# Calcul des coefficients (a et b)
num = np.sum((X - mean_x) * (y - mean_y))  # numérateur
den = np.sum((X - mean_x)**2)              # dénominateur
a = num / den                              # calcul de la pente
b = mean_y - (a * mean_x)                  # calcul de l'intercepte

print(f"Équation scratch: y = {a:.2f}x + {b:.2f}")

# Prédiction
y_pred = a * X + b  # calcul manuel de la droite

# Graphique
pt.scatter(X, y, color='blue')  # nuage de points
pt.plot(X, y_pred, color='red')  # ligne de régression
pt.xlabel("Heures")
pt.ylabel("Note")
pt.title("Modèle fait maison (Scratch)")
pt.show()
```
